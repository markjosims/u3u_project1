# Training
Train Decision Tree classifier on several subsets of features

```python
import sklearn
from glob import glob
import pandas as pd
```

```python
train_csv = 'data/dev/dev-clean-whisper-large-v2-hypotheses-preprocessed.csv'
train_df = pd.read_csv(train_csv)
# train_df = train_df.reindex(sorted(train_df.columns), axis=1)
test_csv = 'data/test/test-clean-whisper-large-v2-hypotheses-preprocessed.csv'
test_df = pd.read_csv(test_csv)
# test_df = test_df.reindex(sorted(test_df.columns), axis=1)

train_df
```

```python
test_df
```

```python
feat_cols = [col for col in train_df.columns if col!='edit_type' and 'POS' not in col]
pos_feats = [col for col in feat_cols if 'pos' in col]
len_feats = [col for col in feat_cols if 'len' in col]
freq_feats = [col for col in feat_cols if 'freq' in col]
identity_feats = [col for col in feat_cols if 'same' in col]
prev_word_feats = [col for col in feat_cols if col.startswith('prev')]
curr_word_feats = [col for col in feat_cols if col.startswith('word')]
next_word_feats = [col for col in feat_cols if col.startswith('next')]
```

```python
feat_groups = {
    "all": feat_cols,
    "pos": pos_feats,
    "len": len_feats,
    "frequency": freq_feats,
    "identity": identity_feats,
    "prev_word": prev_word_feats,
    "curr_word": curr_word_feats,
    "next_word": next_word_feats,
}
```

```python
y = train_df['edit_type'].to_numpy()
y_test = test_df['edit_type'].to_numpy()

for feat_name, feats in feat_groups.items():
    print(feat_name)
    tree_clf = sklearn.tree.DecisionTreeClassifier(max_depth=40)
    X = train_df[feats].to_numpy()
    X_test = test_df[feats].to_numpy()
    tree_clf.fit(X, y)
    y_hat = tree_clf.predict(X_test)
    f1_score = sklearn.metrics.f1_score(y_test, y_hat, average=None)
    print(f1_score)
    
```
